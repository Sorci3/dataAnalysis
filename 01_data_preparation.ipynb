Après avoir vérifier la distribution des deux classe de notre dataset on constate qu'il y à beaucoup plus de personnes ayant remboursé leurs prets (classe de 0) que l'inverse (classe de 1). Cela implique que nous allons devoir prêter une attention perticulière à ce déséquilibre pour le bon entraînement de notre modèle, notammant en leur appliquant des poids différents par exemple.

La suite logique est l'exploration de nos données en vérifiant la présence de valeurs manquantes par exemple, ou bien de valeurs aberrantes. 

```python
def missing_values_report(df):
        # Total missing values
        missing_values = df.isnull().sum()
        
        # Percentage of missing values
        missing_values_percent = (missing_values/len(df))*100
        
        # Make a table with the results
        mis_val_table = pd.concat([missing_values, missing_values_percent], axis=1)
        mis_val_table.columns = ['Missing Values', 'Percentage']

        mis_val_table = mis_val_table[mis_val_table['Missing Values'] > 0]
    
        # Sort by ascending order
        mis_val_table = mis_val_table.sort_values(by='Missing Values', ascending=False)
    
        return mis_val_table
```

```python
missing_values = missing_values_report(train_data)
missing_values.head(50)
```
